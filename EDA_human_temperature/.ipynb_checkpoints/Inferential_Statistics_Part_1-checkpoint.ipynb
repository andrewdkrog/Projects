# What is the True Normal Human Body Temperature? 

#### Background

The mean normal body temperature was held to be 37$^{\circ}$C or 98.6$^{\circ}$F for more than 120 years since it was first conceptualized and reported by Carl Wunderlich in a famous 1868 book. But, is this value statistically correct?

```python
import pandas as pd
import numpy as np
import statsmodels

df = pd.read_csv('data/human_body_temperature.csv')
```

### Is the distribution of human body temperatures normal?

```python
df.head()
```

#### Test normality assumption visually

```python
# Quantile-Quantile plot
%matplotlib inline

from statsmodels.graphics.gofplots import qqplot
from matplotlib import pyplot as plt

qqplot(df.temperature, line='s')
plt.show()
```

```python
# Define ecdf function to create sample cumulative distribution

def ecdf(data):
    """Compute ECDF for a one-dimensional array of measurements."""

    # Number of data points: n
    n = len(data)

    # x-data for the ECDF: x
    x = np.sort(data)

    # y-data for the ECDF: y
    y = np.arange(1, n+1) / n

    return x, y
```

```python
# Calculate mean and standard deviation
mean_temp = np.mean(df.temperature)
std_temp = np.std(df.temperature)
temp_samples = np.random.normal(mean_temp, std_temp, size=10000)

x_temp, y_temp = ecdf(df.temperature)
x_theor, y_theor = ecdf(temp_samples)
```

```python
import seaborn as sns
sns.set()


_ = plt.plot(x_theor, y_theor)
_ = plt.plot(x_temp, y_temp, marker='.', linestyle='none')
_ = plt.xlabel('Body Temperature (F)')
_ = plt.ylabel('CDF')
_ = plt.legend(['Theoretical Normal CDF', 'Sample CDF'])
plt.margins(0.02)
plt.show()
```

#### Test normality assumption using normaltest from scipy.stats package

```python
from scipy.stats import normaltest

# normality test
stat, p = normaltest(df.temperature)
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
```

##### Bootstrap Hypothesis Test

```python
# Define function to compute single bootstrap sample with flexible choice of function to apply to sample
def bootstrap_replicate_1d(data, func):
    bs_sample = np.random.choice(data, size=len(data))
    
    return func(bs_sample)
```

```python
# Define function to iterate through bootstrap replicates
def draw_bs_reps(data, func, size=1):
    """Draw bootstrap replicates."""

    # Initialize array of replicates: bs_replicates
    bs_replicates = np.empty(size)

    # Generate replicates
    for i in range(size):
        bs_replicates[i] = bootstrap_replicate_1d(data, func)

    return bs_replicates
```

```python
# Create a vector of bootstrap means using 10,000 replicates
bs_replicates = draw_bs_reps(df.temperature, np.mean, size = 10000)
```

```python
# Compute fraction of replicates that are less than the observed Frog B force: p
p = np.sum(bs_replicates >= 98.6) / len(bs_replicates)

# Print the p-value
print('p = ', p)
```

```python
# Compute the 95% confidence interval
conf_int = np.percentile(bs_replicates, [2.5, 97.5])

# Print the confidence interval
print('95% confidence interval =', conf_int, 'degrees Fahrenheit')
```

The 95% confidence interval does not contain 98.6 degrees Fahrenheit. It seems likely (as long as there are issues with the representativeness of our sample) that the true population mean of human body temperature is below 98.6 degrees.

##### Frequentist Test

For the frequentist test, the one-sample test will be used since the mean of the sample will be compared against a known constant, 98.6 degrees. We will use the t-statistic rather than z because the population variance is unknown, we only have a sample.

```python
# One-sample t-test
from scipy.stats import ttest_1samp
t_stat, p_val_t = ttest_1samp(df.temperature, 98.6)

print('t-statistic:', t_stat)
print('p value =', p_val_t)
```

```python
# For fun to compare, the z-statistic is calculated which assumes the normal distribution with known population variance
from statsmodels.stats.weightstats import ztest
z_stat, p_val_z = ztest(df.temperature, value = 98.6)

print('z-statistic:', z_stat)
print('p value =', p_val_z)
```

Either way, the null hypothesis that the population mean equals 98.6 degrees Fahrenheit is rejected.
The p-value for the z-statistic is more extreme because of the underlying assumption of normal distribution/known variance.

### Draw a small sample of size 10 from the data and repeat both frequentist tests.

```python
# Take a random sample of size 10 from the 130 observations of body temperature. Do not use replacement.
sample = np.random.choice(df.temperature, size = 10, replace = False)
```

```python
# One-sample t-test
from scipy.stats import ttest_1samp
t_stat, p_val_t = ttest_1samp(sample, 98.6)

print('t-statistic:', t_stat)
print('p value =', p_val_t)
```

```python
# For fun to compare, the z-statistic is calculated which assumes the normal distribution with known population variance
from statsmodels.stats.weightstats import ztest
z_stat, p_val_z = ztest(x1 = sample, value = 98.6)

print('z-statistic:', z_stat)
print('p value =', p_val_z)
```

Using a random sample of size 10, the p-values are no longer significant. The sample is not large enough to establish that the population mean is significantly different from 98.6 degrees.

### At what temperature should we consider someone's temperature to be "abnormal"?

- Start by computing the margin of error and confidence interval. When calculating the confidence interval, keep in mind that you should use the appropriate formula for one draw, and not N draws.

##### Bootstrap

```python
# Create a bootstrap sample of 10,000 observations
sample_size = 10000
bs_sample = np.random.choice(df.temperature, size = sample_size)

# Compute the 95% confidence interval from the bootstrap sample
conf_int = np.percentile(bs_replicates, [2.5, 97.5])
```

```python
print('The bootstrapped 95% confidence interval is ',conf_int[0], 
      ' to ',conf_int[1])
```

##### Frequentist

```python
from scipy import stats

# Calculate number of observations
n=len(df.temperature)
std_temp = np.std(df.temperature)

# Calculate standard error
std_error = std_temp / np.sqrt(n)
print('The standard error is %s' %std_error)

# Calculate 95% confidence interval
conf_int = stats.t.interval(alpha = .95, df = n-1, loc = mean_temp,
                 scale = stats.sem(df.temperature))
print('The 95% confidence interval is from ',conf_int[0], 
      ' to ',conf_int[1])
```

### Is there a significant difference between males and females in normal temperature?

```python
# Use two-sample t-test to test whether there is a difference
male_temps = df.temperature[df.gender == 'M']
female_temps = df.temperature[df.gender == 'F']

print('Mean body temp of male samples: ', np.mean(male_temps))
print('Mean body temp of female samples: ', np.mean(female_temps))

gender_ttest, gender_pval = stats.ttest_ind(a=male_temps, b=female_temps, equal_var=False)
print('The T-Statistic is %d with a p-value of %s' %(abs(gender_ttest), gender_pval))
```

The t-test has a p-value of .024 so we reject the null hypothesis that there is no significant difference between male and female body temperatures.
